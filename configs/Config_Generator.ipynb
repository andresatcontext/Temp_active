```python
import os
import yaml

# total images for training
#config["NUM_TRAIN"] = len(x_train) # N
## shape input
#config["input_shape"] = x_train.shape[1:]
config = {}


#############################################################################################
# PROJECT CONFIG
#############################################################################################
config["PROJECT"] = {}
config["PROJECT"]['project'] = "Active_Learning_CIFAR"
config["PROJECT"]["source"]  = "CIFAR" # DataNet
config["PROJECT"]["dataset_name"]  = "CIFAR" # Name of the dataset in DataNet or test cifar
config["PROJECT"]["project_dir"] = "/mnt/Ressources/Andres/Temp_active/runs"
config["PROJECT"]["Backbone"] = 'resnet18'
config["PROJECT"]["group"] = "CIFAR_LOSSNET"
config["PROJECT"]['group_dir'] = os.path.join(config["PROJECT"]["project_dir"], config["PROJECT"]["group"])


#############################################################################################
# RUNS CONFIG
#############################################################################################
# Parameters of the running proyect
config["RUNS"] = {}
# 

# how many images infer to check its importance to training
config["RUNS"]["SUBSET"] = 10000 # -1 to check all the available images
# from the subset select the best ADDENDUM images to train the network
config["RUNS"]["ADDENDUM"] = 1000 # K

# How many times train test the algorithm with different starting points (different data to train 0)
config["RUNS"]["TRIALS"] = 1
# for how many cycles for annotation time (CYCLES*ADDENDUM"] = total labeled images)
config["RUNS"]["CYCLES"] = 10

config["RUNS"]["test_each"] = 5 
#############################################################################################
# ACTIVE_L CONFIG
#############################################################################################
# Parameters of the running proyect
config["ACTIVE_ALGO"] = {}
# 
config["ACTIVE_ALGO"]["LOSSLEARNING"] = 1.0
# 
#config["ACTIVE_ALGO"]["CORESET"] = 1.0
# 
#config["ACTIVE_ALGO"]["ENTROPY"] = 1.0

#############################################################################################
# DATASET
#############################################################################################
config["DATASET"] = {}
# total images for training
#config["DATASET"]["num_images"] = None# len(x_train) # N

# parametres data augmentation
config["DATASET"]["width_shift_range"] = 4
config["DATASET"]["height_shift_range"] = 4
config["DATASET"]["horizontal_flip"] = True

#############################################################################################
# DATASET
#############################################################################################
# Parameters network
config["NETWORK"] = {}
# shape input
config["NETWORK"]["INPUT_SIZE"] = None # x_train.shape[1:]
# number of classes
config["NETWORK"]["CLASSES"] = None# len(np["unique(y_train))
# length embedding for z
config["NETWORK"]["embedding_size"] = 128
# Margin for the Loss of LossNet
config["NETWORK"]["MARGIN"] = 1.0


#############################################################################################
# DATASET
#############################################################################################
# Parameters Train
config['TRAIN'] = {}
# 
config['TRAIN']["batch_size"] = 128
#
config['TRAIN']["transfer_weight_path"] = False
#
config['TRAIN']["start_epoch"] = 0
# Data augementation
config['TRAIN']["Data_augementation"] = True
# Parametes optimizer
# learning rate
config['TRAIN']["lr"] = 0.001
# Gamma for MultiStepLR
config['TRAIN']["gamma"] = 1e-1
#
config['TRAIN']["wdecay"] = 0.9995
# Parameters losses
#config['TRAIN']["reduction"] ='mean' # 'none'
# Warm up epochs
config['TRAIN']["EPOCH_WARMUP"] = 2
# whole training epochs
config['TRAIN']["EPOCH_WHOLE"] = 40
# split training epochs
config['TRAIN']["EPOCH_SLIT"] = 10
# change learning rate after the numbers in the list
config['TRAIN']["MILESTONES"] = [160]
# weights when adding the losses
config['TRAIN']["w_c_loss"] = 1.0
config['TRAIN']["w_l_loss"] = 1.0


#############################################################################################
# TEST
#############################################################################################
# Parameters test
config['TEST'] = {}
#
config['TEST']["batch_size"] = 128
# Parameters losses
#config['TEST']["reduction"] ='none' # 'none'


#############################################################################################
# DATASET
#############################################################################################
train_with_all_the_data = False

if train_with_all_the_data:

    # how many images infer to check its importance to training
    config["RUNS"]["SUBSET"] = -1 # -1 to check all the available images
    # How many times train test the algorithm with different starting points (different data to train 0)
    config["RUNS"]["TRIALS"] = 1
    # for how many cycles for annotation time (CYCLES*ADDENDUM"] = total labeled images)
    config["RUNS"]["CYCLES"] = 1
    
    
    # Warm up epochs
    config['TRAIN']["EPOCH_WARMUP"] = 2
    # whole training epochs
    config['TRAIN']["EPOCH_WHOLE"] = 40
    # split training epochs  
    #stop the gradient from the loss prediction module propagated to the target model
    config['TRAIN']["EPOCH_SLIT"] = 10
    # change learning rate after the numbers in the list
    config['TRAIN']["MILESTONES"] = [25, 35]
    # weights when adding the losses
    config['TRAIN']["w_c_loss"] = 1.0
    config['TRAIN']["w_l_loss"] = 0.0


else:

    # how many images infer to check its importance to training
    config["RUNS"]["SUBSET"] = -1 # -1 to check all the available images
    # How many times train test the algorithm with different starting points (different data to train 0)
    config["RUNS"]["TRIALS"] = 1
    # for how many cycles for annotation time (CYCLES*ADDENDUM"] = total labeled images)
    config["RUNS"]["CYCLES"] = 10

    
    # Warm up epochs
    config['TRAIN']["EPOCH_WARMUP"] = 2
    # whole training epochs
    config['TRAIN']["EPOCH_WHOLE"] = 120
    # split training epochs
    config['TRAIN']["EPOCH_SLIT"] = 80
    # change learning rate after the numbers in the list
    config['TRAIN']["MILESTONES"] = [160]
    # weights when adding the losses
    config['TRAIN']["w_c_loss"] = 1.0
    config['TRAIN']["w_l_loss"] = 1.0

Config_file_name = 'AL_cifar.yml'

with open(Config_file_name, 'w') as file:
    documents = yaml.dump(config, file)
```
